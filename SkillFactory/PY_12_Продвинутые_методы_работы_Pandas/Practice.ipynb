```python
import pandas as pd
```

```python
raiting = pd.read_csv('data/ratings_movies.csv', index_col='Unnamed: 0')
display(raiting)
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
raiting['year_release'] = raiting['title'].apply(get_year_release)
display(raiting)
```

```python
'''Какой фильм, выпущенный в 1999 году, получил наименьшую среднюю оценку зрителей?
В качестве ответа запишите название этого фильма без указания года его выпуска.'''
raiting[raiting['year_release'] == 1999.0].groupby('title')['rating'].mean().sort_values()
```

```python
#Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, получило наименьшую среднюю оценку (rating)?
raiting[raiting['year_release'] == 2010.0].groupby('genres')['rating'].mean().sort_values()
```

```python
raiting.groupby('userId')['genres'].nunique().sort_values()
```

```python
#Найдите пользователя, который выставил наименьшее количество оценок, но его средняя оценка фильмам наибольшая.
raiting.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(['count', 'mean'], ascending=[True, False])
```

```python
'''Найдите сочетание жанров (genres) за 2018 году, 
которое имеет наибольший средний рейтинг (среднее по столбцу rating), 
и при этом число выставленных ему оценок (количество значений в столбце rating) больше 10.'''
temp = raiting[raiting['year_release'] == 2018].groupby('genres')['rating'].agg(['count', 'mean'])
temp[temp['count'] > 10].sort_values(by = 'mean', ascending=False)
```

```python
raiting['date'] = pd.to_datetime(raiting['date'])
```

```python
raiting['year_rating'] = raiting['date'].dt.year
```

```python
'''Добавьте в таблицу новый признак year_rating — год выставления оценки. 
Создайте сводную таблицу, которая иллюстрирует зависимость 
среднего рейтинга фильма от года выставления оценки и жанра. 
Выберите верные варианты ответа, исходя из построенной таблицы:'''
raiting_pivot = raiting.pivot_table(values='rating', index='genres', columns='year_rating', aggfunc='mean')
```
